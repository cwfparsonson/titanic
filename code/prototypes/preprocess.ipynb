```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Useful Functions

Define any useful functions

```python
def display_all(df,max_rows=1000,max_cols=1000):
    '''
    Displays all data in dataframe with analysis of var stats
    '''
    with pd.option_context('display.max_rows',max_rows,'display.max_columns',max_cols):
        return df
```

## Load Data
Load the data as pandas dataframe

```python
# LOAD DATA
files_dir = '../../files/'
train_data = pd.read_csv(files_dir+'train.csv')
test_data = pd.read_csv(files_dir+'test.csv')
print('Train data: {}'.format(train_data.shape))
print('Test data: {}'.format(test_data.shape))

# INITIALISE DATAFRAME
df = pd.concat([train_data, test_data], axis=0, sort=True)
headers = df.columns.values.tolist()
print('Data info:')
df.info()
print('Headers:\n{}'.format(headers))
print('Initial DataFrame:\n{}'.format(df))
```

### Clean Up Variables

Some parameters have empty/NaN values. ML training cannot handle NaN values as will crash loss func evaluation/make loss NaN. 3 main possible ways to deal with this:

1. Remove any sets of data with a NaN/empty parameter value. Not recommended as will waste a lot of data that would otherwise be helpful for training ML model.

2. Negate any NaN/empty data by setting NaN to e.g. 0 or -1. Also not recommended as the meaning of having 0 or -1 when training ML model, when plotting data etc may be misleading to both the ML model and also to readers.

3. For continuous vars, replace NaN/missing values with mean, median or mode values for that particular variable. If have some outliers, use median to reduce anomaly effect. For categorical vars, replace with most common class. This is the common standard approach to dealing with missing values.

4. Use regression and/or another ML model; use the other parameter values for this set of data (e.g. sex, class, title etc) to try to predict what the variable value would have been

For our dataset, there are 177 ages missing and 2 fares missing.

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

Can see that there are differences in survival rate depending on both age and sex. Men are more likely to surive between 18-30 years old, whereas for women the likley survival range is 14-40. 5-18 year olds have low survival chances, whereas infants have a higher survival chance. 

We can conclude that certain age groups have different chances of survival, and that both age and sex will be important for our algorithm

```python
# Fare

# fare is continuous, but would be better to categorise so get fewer possible values.
# Can categorise into 6 classes

df.loc[ df['Fare'] <= 7.91, 'Fare'] = 0
df.loc[(df['Fare'] > 7.91) & (df['Fare'] <= 14.454), 'Fare'] = 1
df.loc[(df['Fare'] > 14.454) & (df['Fare'] <= 31), 'Fare']   = 2
df.loc[(df['Fare'] > 31) & (df['Fare'] <= 99), 'Fare']   = 3
df.loc[(df['Fare'] > 99) & (df['Fare'] <= 250), 'Fare']   = 4
df.loc[ df['Fare'] > 250, 'Fare'] = 5
df['Fare'] = df['Fare'].astype(int)

plt.figure()
axes = sns.factorplot('Fare', 'Survived', data=df, aspect=2.5)
plt.show()
```

There is a clear relationship between fare and survival rate

## Process the Variables

There are two types of variables:

1. Categorical variables
2. Continuous variables

We must decide which variables we want to keep, and we must convert them into ML-readable formats.

Vars in original data:
1. 'Age' 
2. 'Cabin' (Cabin number)
3. 'Embarked' (Port of embarkation)
4. 'Fare' 
5. 'Name' (Full name and title)
6. 'Parch' (# of parents & children aboard)
7. 'PassengerId' 
8. 'Pclass' (Ticket class
9. 'Sex' 
10. 'SibSp' (# of siblings & spouses aboard)
11. 'Survived'
12. 'Ticket' (Ticket #)

Categoircal vars we want to keep (binarise or 1-hot encode):
1. 'Sex': Binary categorical therefore encode as single 1-0 column
2. 'Embarked': Multivariate categorical therefore 1-hot encode
3. 'Fare': After our transformation, is multivariable cat (not cont), therefore 1-hot encode

Continuous vars we want to keep (must be standardised w/ centering and scaling):
1. 'Age'
3. 'Parch'
4. 'SibSP'


```python
# CATEGORICAL VARS

# sex
df['Sex'] = df['Sex'].astype('category') # conv to category dtype
df['Sex'] = df['Sex'].cat.codes # conv to 1s & 0s

# embarked
embarked_encoded = pd.get_dummies(df['Embarked'], prefix='Embarked') # one hot encode
df = pd.concat([df, embarked_encoded], axis=1) # append encoded embarked
del df['Embarked'] # remove original embarked

# fare
fare_encoded = pd.get_dummies(df['Fare'], prefix='Fare')
df = pd.concat([df, fare_encoded], axis=1)
del df['Fare']

df.drop(['Cabin', 'Name', 'Ticket', 'PassengerId'], axis=1, inplace=True)
print('Removed remaining categorical variables we don\'t want')
print(df)
```

```python
# CONTINUOUS VARS

cont_vars = ['Age', 'Parch', 'Pclass', 'SibSp']
scaler = StandardScaler()
for var in cont_vars:
    df[var] = df[var].astype('float64') # conv data to float
    df[var] = scaler.fit_transform(df[var].values.reshape(-1,1)) # center data and transform
print(df)
```

```python
# Display Processed Data

print('Processed Data:\n{}'.format(df))
print('Analysis:\n{}'.format(display_all(df.describe(include='all').T)))
print('Headers:\n{}'.format(df.columns.values.tolist()))
print('Shape: {}'.format(df.shape))    
```
